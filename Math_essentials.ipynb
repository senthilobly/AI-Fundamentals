### Dot Product Explained

The **dot product** of two vectors is a way to multiply them and get a **scalar** (a single number) as a result.

### Dot Product Formula (Algebraic and Geometric)

Let **a** and **b** be two vectors:

**Algebraic form**:
vec{a} dot vec{b} = a_1b_1 + a_2b_2 + dots + a_n b_n = sum_{i=1}^{n} a_i b_i

**Geometric form**:
vec{a} dot vec{b} = |vec{a}| dot | vec{b}| dot cos(theta)

Where:
- (vec{a},vec{b}) are vectors
- (a_i, b_i) are the components of each vector
- (|vec{a}|) is the magnitude (length) of vector a
- (theta) is the angle between the vectors


```python
# Dot product
v1 = np.array([1, 2])
v2 = np.array([3, 4])
dot = np.dot(v1, v2)
print("Dot product of v1 and v2 :", dot)
```

## Matrix Multiplication?

Matrix multiplication is a way of combining two matrices to produce a new matrix.

It’s **not** like element-wise multiplication. Instead, each element in the result matrix is calculated by taking the **dot product** of a row from the first matrix and a column from the second matrix.

If matrix **A** is of size(m times n), and matrix **B** is of size(n times p),  
then the matrix product **C = AB** will be of size(m times p).

Matrix Calculaiton = C_{ij} = sum_{k=1}^{n} A_{ik} dot B_{kj}

```python
# Matrix Multiplication
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
mat_mul = np.dot(A,B)
print("Matrix multiplication of A and B:", mat_mul)
```

## Matrix Transpose and Inverse Operations

---

### 1. Matrix Transpose

The **transpose** of a matrix is obtained by flipping the matrix over its diagonal.  
This means **rows become columns** and **columns become rows**.

If:

A =  
[  
 a₁₁ a₁₂  
 a₂₁ a₂₂  
]  
then:  
Aᵀ =  
[  
 a₁₁ a₂₁  
 a₁₂ a₂₂  
]

#### Properties of Transpose:
- (Aᵀ)ᵀ = A  
- (A + B)ᵀ = Aᵀ + Bᵀ  
- (AB)ᵀ = Bᵀ Aᵀ  

---

### 2. Matrix Inverse

The **inverse** of a square matrix A is another matrix A⁻¹ such that:

A × A⁻¹ = A⁻¹ × A = I

Where:
- I is the identity matrix  
- Only square matrices (same number of rows and columns) that are **non-singular** (determinant ≠ 0) have an inverse.

---

### Inverse Formula for 2×2 Matrix:

If:

A =  
[  
 a b  
 c d  
]  
then:  

A⁻¹ = (1 / (ad − bc)) ×  
[  
 d −b  
 −c a  
]

*Condition:* ad − bc ≠ 0 (determinant ≠ 0)

---

### Example:

Let:  
A =  
[  
 1 2  
 3 4  
]

Then:

- **Transpose**:  
Aᵀ =  
[  
 1 3  
 2 4  
]

- **Inverse**:  
det(A) = (1×4 − 2×3) = 4 − 6 = −2  

A⁻¹ = (1 / −2) ×  
[  
 4 −2  
 −3 1  
]  
=  
[  
 −2.0 1.0  
 1.5 −0.5  
]


```python
# Transpose Operation
A = np.array([[1,2],[3,4]])
transpose = A.T
print("A transpose :", transpose)
```

```python
# Inverse Operation
inverse = np.linalg.inv(A)
print("Inverse of A:", inverse)
```

## Derivative and Gradient Explained

---

## Derivative in Deep Learning

---

### What is a Derivative?

A **derivative** is a measure of **how much a function changes** when its input changes — in simple terms, it’s the **rate of change**.

In deep learning, derivatives are used to:
- Understand how changes in weights and biases affect the **loss**
- Update weights to **reduce error**

---

### Mathematical Definition

\[
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
\]

We can't actually compute the limit in code, so we use a very small **h**, like **1e-5**, to approximate it.

---

### Example: Derivative of f(x) = x²

\[
f'(x) = 2x
\]

At **x = 3**:  
\[
f'(3) = 6
\]

Which means the slope (rate of change) of the function at x = 3 is 6.

---

## Gradient and Gradient Descent in Deep Learning

---

### 1. What is a Gradient?

A **gradient** is a vector of partial derivatives.

In multivariable functions, it tells us:
- **How each input variable affects the output**
- **In which direction** to change the inputs to **increase or decrease** the function value

For a function:  
**f(x, y) = x² + y²**

The gradient is:  
**∇f(x, y) = [ ∂f/∂x , ∂f/∂y ] = [2x, 2y]**

---

### What the Gradient Tells Us

- It points in the **direction of steepest increase** of the function
- If we want to **minimize** the function, we move in the **opposite direction**
- In deep learning, we apply this to **loss functions** to improve model weights

---

### 2. Gradient Calculation Example

Let:

**Loss = (y - wx)²**  
(where w is a weight)

We want to compute the derivative of Loss w.r.t weight **w**:

1. Expand the loss:  
 **L = (y - wx)²**

2. Take derivative:  
 **dL/dw = 2(wx - y) * x**

This gives the **direction and magnitude** in which weight `w` affects the loss.

---

### 3. What is Gradient Descent?

**Gradient Descent** is an optimization algorithm used to **minimize** loss functions by updating weights.

It updates each weight using:

\[
w = w - η \cdot \frac{∂L}{∂w}
\]

Where:
- \( w \) is the weight
- \( η \) is the learning rate (step size)
- \( \frac{∂L}{∂w} \) is the gradient of loss w.r.t the weight

---

### Gradient Descent Steps:

1. Initialize weights randomly
2. For each epoch:
   - Compute the prediction
   - Compute the loss
   - Compute the gradient (partial derivatives)
   - Update the weights using:  
  **w = w - η * gradient**
3. Repeat until loss is small or convergence is reached

---

### Summary Table

| Concept           | Meaning                                                  |
|------------------|-----------------------------------------------------------|
| Gradient          | Vector showing rate of change for each input variable    |
| Partial Derivative| Derivative w.r.t. one variable while others stay fixed   |
| Gradient Descent  | Method to minimize loss by adjusting weights             |
| Learning Rate (η) | Controls step size in gradient descent                   |

---

```python
# Derivative calculation
def f(x):
    return x ** 2

def derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

x_val = 3
print("Approximate derivative at x = 3:", derivative(f, x_val))
```

```python
# Simple gradient descent for f(w) = (y - wx)^2

x = 2
y = 10
w = 1.0
learning_rate = 0.1

for epoch in range(20):
    y_pred = w * x
    loss = (y - y_pred)**2
    gradient = 2 * (y_pred - y) * x
    w = w - learning_rate * gradient
    print(f"Epoch {epoch}: w = {w:.4f}, loss = {loss:.4f}")
```

```python
# Target: y = wx, where x = 2 and y_true = 10
x = 2
y_true = 10

# Initialize weight and learning rate
w = 1.0
learning_rate = 0.1
epochs = 50

# Store values for plotting
weights = []
losses = []

# Gradient Descent Loop
for epoch in range(epochs):
    y_pred = w * x
    loss = (y_true - y_pred)**2
    dL_dw = 2 * (y_pred - y_true) * x
    w = w - learning_rate * dL_dw
    weights.append(w)
    losses.append(loss)

# Plotting Loss vs Epoch
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(epochs), losses, 'b-o')
plt.title('Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# Plotting Weight Updates
plt.subplot(1, 2, 2)
plt.plot(range(epochs), weights, 'r-o')
plt.title('Weight Convergence')
plt.xlabel('Epoch')
plt.ylabel('Weight Value')

plt.tight_layout()
plt.grid(True)
plt.show()

# Plotting Loss w.r.t each weight
plt.figure(figsize=(12,5))
plt.subplot(1,1,1)
plt.plot(weights, losses, 'g-o')
plt.title("Losses w.r.t weight")
plt.xlabel('Weight')
plt.ylabel('Loss')
plt.tight_layout()
plt.grid(True)
plt.show()
```

## Derivative of Activation Functions in Deep Learning

---

### Why Are Derivatives Important?

During backpropagation, we use the **derivative of the activation function** to:
- Scale the gradient that flows back
- Decide how much to update weights

If the derivative is **too small**, learning slows (vanishing gradient).  
If it’s **zero**, the neuron is "dead" (especially in ReLU).

---

###  Common Activation Functions and Their Derivatives

---

### 1. Sigmoid

**Function:**  
σ(x) = 1 / (1 + e^(-x))

**Derivative:**  
σ'(x) = σ(x) * (1 - σ(x))

---

### 2. Tanh

**Function:**  
tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))

**Derivative:**  
d/dx tanh(x) = 1 - tanh²(x)

---

### 3. ReLU

**Function:**  
f(x) = max(0, x)

**Derivative:**  
f'(x) =  
  1 if x > 0  
  0 if x ≤ 0

---

### 4. Leaky ReLU

**Function:**  
f(x) =  
  x if x > 0  
  αx if x ≤ 0

**Derivative:**  
f'(x) =  
  1 if x > 0  
  α if x ≤ 0

---

### Tip:
- Use Sigmoid/Tanh in the output or shallow nets
- Use ReLU/LeakyReLU in hidden layers to avoid vanishing gradients

---


```python
x = np.linspace(-10, 10, 400)

# Sigmoid and its derivative
sigmoid = 1 / (1 + np.exp(-x))
sigmoid_derivative = sigmoid * (1 - sigmoid)

# Tanh and its derivative
tanh = np.tanh(x)
tanh_derivative = 1 - tanh**2

# ReLU and its derivative
relu = np.maximum(0, x)
relu_derivative = np.where(x > 0, 1, 0)

# Leaky ReLU and its derivative
alpha = 0.1
leaky_relu = np.where(x > 0, x, alpha * x)
leaky_relu_derivative = np.where(x > 0, 1, alpha)

# Plot all functions and derivatives
fig, axs = plt.subplots(4, 2, figsize=(12, 12))

# Sigmoid
axs[0, 0].plot(x, sigmoid, label='Sigmoid')
axs[0, 1].plot(x, sigmoid_derivative, label="Sigmoid'", color='r')
axs[0, 0].set_title("Sigmoid")
axs[0, 1].set_title("Sigmoid Derivative")

# Tanh
axs[1, 0].plot(x, tanh, label='Tanh')
axs[1, 1].plot(x, tanh_derivative, label="Tanh'", color='r')
axs[1, 0].set_title("Tanh")
axs[1, 1].set_title("Tanh Derivative")

# ReLU
axs[2, 0].plot(x, relu, label='ReLU')
axs[2, 1].plot(x, relu_derivative, label="ReLU'", color='r')
axs[2, 0].set_title("ReLU")
axs[2, 1].set_title("ReLU Derivative")

# Leaky ReLU
axs[3, 0].plot(x, leaky_relu, label='Leaky ReLU')
axs[3, 1].plot(x, leaky_relu_derivative, label="Leaky ReLU'", color='r')
axs[3, 0].set_title("Leaky ReLU")
axs[3, 1].set_title("Leaky ReLU Derivative")

# Format all subplots
for ax_row in axs:
    for ax in ax_row:
        ax.grid(True)
        ax.legend()

plt.tight_layout()
plt.show()
```

# Probability and Statistics for Deep Learning  
## Mean, Variance, and Standard Deviation

---

## 1. Mean (Average)

### ➤ Definition:
The **mean** is the central (average) value of a dataset.

**Formula:**
Mean (μ) = (x₁ + x₂ + ... + xₙ) / n

- n = number of data points  
- xᵢ = each individual value

### ➤ Use in Deep Learning:
- Calculating average **loss**
- Summarizing weights or activations
- Used in **Batch Normalization**

```python
data = [2, 4, 6, 8, 10]
mean = np.mean(data)
print("Mean:", mean)
```

## Variance in Deep Learning

### ➤ What is Variance?

Variance tells us **how much the values in a dataset differ from the mean**.  
In Deep Learning, it is used to understand the **spread of data**, **weights**, or **activations**, and is a key part of **normalization techniques** like **Batch Normalization**.

---

### Why Does It Matter in Deep Learning?

- High variance in inputs or weights can cause **unstable training**.
- Normalizing inputs (mean = 0, variance = 1) improves **gradient flow** and **convergence**.
- Variance is used in **weight initialization** strategies like Xavier or He initialization.


### Simple Formulas

σ² = [(x₁ - μ)² + (x₂ - μ)² + ... + (xₙ - μ)²] / n

xᵢ: each data point
μ: population mean (Value calculated form Mean for the given Data)
n: total number of data points


```python
variance = np.var(data)
print("Variance:", variance)
```

## Standard Deviation – With Formula and Example

### ➤ What is Standard Deviation?

Standard deviation measures **how spread out the values** in a dataset are from the mean.  
It is the **square root of the variance**, so it's in the same unit as the data.

---

### Formula for Population Standard Deviation

σ = √[ ( (x₁ - μ)² + (x₂ - μ)² + ... + (xₙ - μ)² ) / n ]

- σ: population standard deviation  
- μ: population mean  
- n: number of data points  
- xᵢ: each data point



```python
For simplicity Standard Deviation SD = Sqrt(Variance)
```

```python
sd = np.std(data)
print("Standar Deviation:", sd)
```
